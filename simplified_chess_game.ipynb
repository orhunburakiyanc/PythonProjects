```python
#Create horizontal,vertical,special characters and alphabet strings.
horizontal="a b c d e f g h"
vertical="1 2 3 4 5 6 7 8"
special_characters="- > < ! ' £ # ^ + $ % ½ & / { ( [ ) ] = } ? \ * _ | , ; . : @ é " 
alphabet="a b c d e f g h i j k l m n o p q r s t u v w x y z"

#Control the inputs 
while(1):
  knighth=input("Please enter horizontal position of the knight (a,b,c,d,e,f,g,h): ").lower()
  if(knighth.isdigit()==True or len(knighth)>=2 or special_characters.find(knighth)!=-1):
    print("Horizontal input for knight is not a letter")
    break
  elif(horizontal.find(knighth)!=-1):
    knightv=input("Please enter vertical position of the knight (1,2,3,4,5,6,7,8): ")
    if(knightv.isdigit==False or alphabet.find(knightv)!=-1 or special_characters.find(knightv)!=-1):
      print("Vertical input for knight is not a number")
      break
    elif(int(knightv)<0):
          print("Vertical input for knight is not a number")
          break
    elif(vertical.find(knightv)==-1):
      print("Vertical input for knight is not a proper number")
      break
    elif(vertical.find(knightv)!=-1):
      bishoph=input("Please enter horizontal position of the bishop (a,b,c,d,e,f,g,h): ").lower()
      if(bishoph.isdigit()==True or len(bishoph)>=2 or special_characters.find(bishoph)!=-1):
        print("Horizontal input for bishop is not a letter")
        break
      elif(horizontal.find(bishoph)==-1):
        print("Horizontal input for bishop is not a proper letter")
        break

      elif(horizontal.find(bishoph)!=-1):
        bishopv=input("Please enter vertical position of the bishop (1,2,3,4,5,6,7,8): ")
        if(bishopv.isdigit==False or special_characters.find(bishopv)!=-1 or alphabet.find(knightv)!=-1):
          print("Vertical input for bishop is not a number")
          break
        elif(int(bishopv)<0):
          print("Vertical input for bishop is not a number")
          break
        elif(vertical.find(bishopv)==-1):
          print("Vertical input for bishop is not a proper number")
          break
        elif(vertical.find(bishopv)!=-1):
          vertical_list=vertical.split()
          horizontal_list=horizontal.split()        
          if(bishoph==knighth and bishopv==knightv):
            print("They can't be in the same square")
            break
          elif(bishoph==knighth or bishopv==knightv):
            print("Neither of them can attack each other")
            break
          elif(((int(knightv)-2==int(bishopv) or int(knightv)+2==int(bishopv)) and (ord(bishoph)==ord(knighth)+1 or ord(bishoph)==ord(knighth)-1)) or ((int(knightv)-1==int(bishopv) or int(knightv)+1==int(bishopv)) and (ord(bishoph)==ord(knighth)-2 or ord(bishoph)==ord(knighth)+2)) ):
            print("Knight can attack bishop")
            break
          elif (abs((horizontal_list.index(knighth))-(horizontal_list.index(bishoph))) / abs ((vertical_list.index(knightv))-(vertical_list.index(bishopv))) == 1 ):
            print("Bishop can attack knight")
            break
          else:
            print("Neither of them can attack each other")
            break
    elif(horizontal.find(knighth)==-1):
      print("Horizontal input for knight is not a proper letter")
      break
    
```
