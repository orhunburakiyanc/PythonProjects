```python
import random as rnd 
def random_letter (name):
  x=0
  password=""
  while x<3:
    letter=name[(rnd.randint(0,len(name)-1))]
    password=password+letter
    x+=1
  return password

def random_number(password):
  a=0
  while a<4:
    number=str(rnd.randint(0,9))
    password=password+number
    a+=1
  return password

name=input("Name: ").lower()
p=random_letter(name)
print(random_number(p))
```
