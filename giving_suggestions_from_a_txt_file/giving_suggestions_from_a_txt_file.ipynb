```python
inFile=open("turkiye_spotify_data.txt",'r')
txt_file=inFile.readlines()
#Delete first index's values "not relevant"
txt_file.pop(0)

#get a dictionary that contains stream(1 song is given more than one time so you have to sum up all these streams)
def moststream(txt,year,month):
  artist = {}
  max = 0
  for row in txt:
    altered_row=row.replace("\n","").split("\t")
    for element in altered_row:
      if element.find("-") != -1:
        dates_of_element = element.split("-")
        if year ==  dates_of_element[0] and month == dates_of_element[1]:
          #artist = {"trackname" : [stream,artistname,url]}
          if not altered_row[1] in artist:
            artist[altered_row[1]] = [0] 
            artist[altered_row[1]].append(altered_row[2])
            artist[altered_row[1]].append(altered_row[4])
          artist[altered_row[1]][0]+=int(altered_row[3])  
  return_to_main = output_order(artist)

  return return_to_main

#the dict 'artist' has summed up streams but the key's do not continue from most to least
def output_order(artist):

  count = 1
  order = {}
  order_moststream = {}
  
  for trackname in artist:
    order_moststream[trackname] = artist[trackname][0]

  for i in range(len(order_moststream)):
    maxStream = max(order_moststream.values())
    for trackname, stream in order_moststream.items():
      if stream == maxStream:
        maxTrackname = trackname
        order[count] = [stream,maxTrackname,artist[maxTrackname][1],artist[maxTrackname][2]]
        break
    count += 1
    del order_moststream[maxTrackname]
  return order

#Input YEAR 
year=input("Enter the year: ")
while year != "2017" and year !="2018": 
  year=input("Enter the year: ")

#Input MONTH
month=input("Enter the month: ")

#Ask month until user write valid variable 
if year == "2017":
  while int(month) > 12 or int(month) < 1:
    month=input("Enter the month: ")
else:
  while month != "1":
    month=input("Enter the month: ")
if int(month) <10:
  month = "0"+month

order = moststream(txt_file,year,month)
for trackname in order:
  print("NEW SUGGESTION: {}, {} (Total stream number in this month: {})".format(order[trackname][1],order[trackname][2],order[trackname][0]))
  decision = input("Do you want to listen this song (enter either yes or no): ").lower()
  while decision != "yes" and decision !="no":
    decision = input("Do you want to listen this song (enter either yes or no): ").lower()
  if decision == "yes":
    print("Enjoy {}, {}. Here is the url for you: {} ".format(order[trackname][1],order[trackname][2],order[trackname][3]))
    break
  elif decision == "no":
    continue
  
   
        
```
