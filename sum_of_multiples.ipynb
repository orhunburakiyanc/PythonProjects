```python
stop_number=input("Please enter the stop number: ")
if stop_number.isdigit():
  stop_number=int(stop_number)
  if stop_number<1:
    print("Stop number cannot be less then 1!")
  else:
    first_multiple=input("Please enter the first multiple: ")
    second_multiple=input("Please enter the second multiple: ")
    if (first_multiple.isdigit() and second_multiple.isdigit()) or (int(first_multiple)<1 or int(second_multiple)<1):
      first_multiple=int(first_multiple)
      second_multiple=int(second_multiple)
      if first_multiple<1 or second_multiple<1:
        print("Multipliers cannot be less than 1!")
      elif first_multiple==second_multiple:
        print("Multipliers cannot be equal to each other!")
      elif first_multiple>=stop_number or second_multiple>=stop_number:
        print("Multipliers cannot be equal to stop number and more than stop number!")
      else:
        a=0
        sum=0
        while True:
          if a % first_multiple==0 or a % second_multiple==0:
            sum=a+sum
          a+=1
          if a==stop_number:
            break
        print("Sum of all multiplies of {} or {} below {} is {}.".format(first_multiple,second_multiple,stop_number,sum))
    else:
      print("Multipliers must be integers!")
else:
  print("Stop number must be an integers!")
```
